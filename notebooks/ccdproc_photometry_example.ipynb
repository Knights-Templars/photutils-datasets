# Aperture photometry of stars in the Landolt Field SA112-SF1

## The image...

This image was taken in July 2013 at [Minnesota State University Moorhead](http://physics.mnstate.edu)'s [Paul Feder Observatory](http://astronomy.mnstate.edu/Feder_Observatory).

It has been reduced with a pre-pre-release version of https:github.com/astropy/ccdproc; though it contains a pixel-by-pixel uncertainty that should not be taken too seriously. Additional testing and implementation of ccdproc needs to be done before being confident in any of the results it produces.

You can see the notebook with that reduction at: http://nbviewer.ipython.org/gist/mwcraig/06060d789cc298bbb08e

```python
image_hdus = fits.open('../data/SA112-SF1-001R1.fit.gz')
image = image_hdus[0]
uncertainty = image_hdus[1]
mask = image_hdus[2]
```

The stuff about converting the mask to ``bool`` will go away once astropy 0.4.0 is released...and hopefully explicitly setting up the uncertainty will not be necessary in the near future.

```python
ccd_image = CCDData(image.data, mask=array(mask.data, dtype=bool_),
                    unit=image.header['bunit'], meta=image.header)
from astropy.nddata import StdDevUncertainty
ccd_image.uncertainty = StdDevUncertainty(uncertainty.data)
```

## Read coordinates of stars to photometer

footnotes: 

+ Star coordinates and counts came from [AstroImageJ](http://www.astro.louisville.edu/software/astroimagej/)
+ Counts in AstroImageJ were done on an image that had *not* been calibrated. Yes, I know that is wrong, but it was the quickest way to come up with a coordinate list.

```python
from astropy.table import Table
star_list = Table.read('../data/SA112-SF1-ra-dec-list.txt', format='ascii')
star_list.colnames
r_ap, r_in, r_out  = star_list['Source_Radius'][0], star_list['Sky_Rad(min)'][0], star_list['Sky_Rad(max)'][0]
print r_ap, r_in, r_out
```

```python
x_cen = star_list['X(FITS)']   # centroid in pixel coordinates
y_cen = star_list['Y(FITS)']
```

## Add WCS information

```python
im_wcs = wcs.WCS(ccd_image.header)
cen_x, cen_y = array(ccd_image.shape)/2
cen_ra_dec = im_wcs.all_pix2world(cen_x, cen_y, 0)
```

## Perform photometry

Acknowledging that "perform photometry" is an action that means different things to different people,what *I* mean in *this* notebook is aperture photometry with background subtraction done by an annulus centered on the aperture.

Right now this requires three steps:

```python
aperture_fluxes = photutils.aperture_circular(ccd_image, x_cen, y_cen, r_ap)
annulus_fluxes = photutils.annulus_circular(ccd_image, x_cen, y_cen, r_in, r_out)
```

```python
net_flux = aperture_fluxes - annulus_fluxes * (r_ap**2/(r_out**2 - r_in**2))
```

Have to add an extra dimension to the aperture list to get them broadcast properly. Seems a little fiddly (though I understand why it currently behaves that way).

```python
many_ap_fluxes = photutils.aperture_circular(ccd_image, x_cen, y_cen, apertures[:, newaxis])
```

```python
many_ap_fluxes.shape
```

Background hasn't changed, so should be able to subtract that still...as long as I'm careful about the change in aperture radii.

Luckily, I happen to be familiar with numpy array broadcasting because of work on ccdproc :)

In a nutshell: put in a ``newaxis`` in any spot where you want the same set of values to be used for all values of other arrays in that spot.

From above, ``many_ap_fluxes`` has number of apertures as its first index and number of stars for its second index.

For calculating backgrounds I want to use the same ``annulus_fluxes`` for each aperture, so I use ``annulus_fluxes[newaxis, :]`` below. I also need to calculate the area of the aperture, which will be the same for each star, so I use ``apertures[:, newaxis]`` below.

```python
many_backgrounds = annulus_fluxes[newaxis, :] * apertures[:, newaxis]**2 / (r_out**2 - r_in**2)
```

```python
many_net_fluxes = many_ap_fluxes - many_backgrounds
```

```python
many_instrumental_mags = -2.5 * log10(many_net_fluxes)
```

```python
for idx in range(len(star_list)):
    plot(apertures, many_instrumental_mags[:, idx], label='star {}'.format(idx))
xlabel('aperture radius, pixels')
ylabel('-2.5 * log of counts')

# Reverse magnitude axis
orig_range = ylim()
ylim(orig_range[1], orig_range[0])

xlim(apertures[0], apertures[-1])
legend()
```
